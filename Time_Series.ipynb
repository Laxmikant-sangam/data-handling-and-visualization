```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Date": pd.to_datetime(["2022-01-01", "2022-02-01", "2022-03-01", "2022-04-01", "2022-05-01"]),
    "Value": [10, 15, 22, 28, 35]
}
df = pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker="X")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Month": ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"],
    "Sales": [100, 150, 130, 162, 180, 200, 225, 205, 195, 220, 225, 230]
}

df = pd.DataFrame(data)
df.set_index("Month", inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Sales"], marker="o")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Time Series with Seasonality")
plt.grid(True)
plt.xticks(rotation=69)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text = "This is a sample text for creating a word cloud. It contains words with different frequencies."


wordcloud = WordCloud().generate(text)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud")


word_counts = Counter(text.split())


words = list(word_counts.keys())
counts = list(word_counts.values())


plt.subplot(1, 2, 2)
plt.bar(words, counts)
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Word Frequency Histogram")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string

input_string = "This is an example sentence to remove stop words"
result = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result)
```
